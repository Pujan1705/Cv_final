```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,10)
plt.rcParams['font.size']=16
sns.set_style('darkgrid')
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,roc_auc_score
```

```python
data = pd.read_csv('classification.csv')
```

```python
data
```

```python
y_true = data['actual'].values
y_pred = data['predicted'].values
```

```python
fpr,tpr,_ = roc_curve(y_true, y_pred)
auc = roc_auc_score(y_true,y_pred)

plt.plot(fpr,tpr,label="AUC Score = "+str(auc),lw=2)
plt.plot([0, 1], [0, 1], color="orange", lw=2, linestyle="--")
plt.legend(loc=4)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

```python
y_pred[y_pred>=0.8]=1
y_pred[y_pred<0.8]=0
```

```python
print(classification_report(y_true,y_pred))
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(confusion_matrix(y_true,y_pred),annot=True,cmap='Blues_r',fmt='.3g',
            xticklabels=['No Covid (0)','Covid (1)'], yticklabels=['No Covid (0)','Covid (1)'])
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```

```python
#shifting to 0.8 threshold makes accuracy better, but at 0.5 we have lower FP. -good for finance problem

# for health problem - we want low false negatives , so we want high true negative rate = 1 - FPR
# so we want low FPR, so take threshold 0.3
```

```python
y_pred[y_pred>=0.5]=1
y_pred[y_pred<0.5]=0
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(confusion_matrix(y_true,y_pred),annot=True,cmap='Blues_r',fmt='.3g',
            xticklabels=['No Covid (0)','Covid (1)'], yticklabels=['No Covid (0)','Covid (1)'])
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```
